```python
# !pip3 install matplotlib
# !pip3 install numpy
import matplotlib.pyplot as plt
import numpy as np
```

```python
tamanho = np.array([100000, 200000, 400000, 800000, 1600000])
trocaInsertion = np.array([2499180166, 9996489965, 39987819383, 159860431456, 640014680536]) / 1e6  # Divide por 1 milhão

plt.figure(figsize=(10, 6))
plt.plot(tamanho, trocaInsertion, marker="o", linestyle='-', color='b')
plt.grid(True)
plt.xlabel("Tamanho do vetor", fontsize=12)
plt.ylabel("Quantidade de trocas (em milhões)", fontsize=12)

plt.title("Relação entre tamanho do vetor e quantidade de trocas", fontsize=14)
plt.show()
```
